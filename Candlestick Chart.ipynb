```python
import pandas as pd 
import numpy as np 
from pandas.testing import assert_frame_equal
from pandas_datareader import data as web

import mplfinance as mpf

# Import the plotting library
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Define dataframe variable
btc = pd.DataFrame()

# Define stock that will be analysed
asset = 'BTC-USD'
 
# Import data into dataframe
btc = web.DataReader(asset, data_source='yahoo', start='01-01-2017')

btc.head()
```

```python
mpf.plot(btc, 
        type = 'candle',
        style =  'binance',
        title = 'BTC Price Chart',
        ylabel = 'BTC Prices', show_nontrading = True, mav = (50,100,200), 
        volume = True, figratio=(25,12), 
        ylabel_lower='Daily Volume')
```
