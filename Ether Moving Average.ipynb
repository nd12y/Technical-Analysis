```python
import pandas as pd 
import numpy as np 
from pandas.testing import assert_frame_equal
from pandas_datareader import data as web

# Import the plotting library
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Define dataframe variable
eth = pd.DataFrame()

# Define stock that will be analysed
asset = 'ETH-USD'
 
# Import data into dataframe
eth = web.DataReader(asset, data_source='yahoo', start='05-01-2017')

eth.head()
```

```python
eth['MA50'] = eth['Close'].rolling(window=50).mean()
eth['MA100'] = eth['Close'].rolling(window=100).mean()
eth['MA200'] = eth['Close'].rolling(window=200).mean()
```

```python
f, ax = plt.subplots(figsize = (20,8))

eth['Close'].plot(color = 'dodgerblue', lw=2, label = 'Ether price')
eth['MA50'].plot(ax = ax, label = 'MA50', color = 'red')
eth['MA200'].plot(ax = ax, label = 'MA200', color = 'darkorange')

ax.set_ylabel('Ether Price', fontsize=16, fontweight = 'bold')
ax.set_xlabel('Year', fontsize=16, fontweight = 'bold')
ax.set_title('Ethereum with Moving Averages', fontsize=25, fontweight='bold')
ax.legend()
```
